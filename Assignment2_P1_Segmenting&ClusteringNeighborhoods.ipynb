```python
# Import libraries
import pandas as pd
import requests
import lxml.html as lh
```

```python
# URL to Wikipedia page
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

page = requests.get(url)

#Store the page contents at the given URL
doc = lh.fromstring(page.content)

#Parse data between the row elements <tr> where the table class is 'wikitable'
data = doc.xpath('//table[@class="wikitable"]//tr')
```

```python
len(data)
```

```python
#Retrieve the headers from the first row 
table=[]
i=0

for t in data[0]:
    i+=1
    header=t.text_content()
    table.append((header,[]))
table
```

```python
for i in range(1,len(data)):
    j=0
    for d in data[i]:
        table[j][1].append(d.text_content() )
        j+=1
```

```python
df_dict={header:col for (header,col) in table}
df=pd.DataFrame(df_dict)
df.columns=['PostalCode','Borough','Neighborhood']
df
```

### Cleaning data

```python
#Removing newline characters
df = df.replace('\n',' ', regex=True)
#Replace '/' with a ',' as required in the format
df = df.replace('/',',', regex=True)
df
```

### Removing Rows where the Borough is 'Not assigned'

```python
df = df[df.Borough!='Not assigned ']
df = df.reset_index(drop=True)
df
```
